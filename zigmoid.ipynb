```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
def sigmoid_de(z):
    return sigmoid(z) * (1 - sigmoid(z))
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = np.array([[0], [1], [1], [0]])
W1 = np.random.randn(2, 2)
b1 = np.random.randn(1, 2)
W2 = np.random.randn(2, 1)
b2 = np.random.randn(1, 1)
lr = 0.1

z1 = X @ W1 + b1
a1 = sigmoid(z1)

z2 = a1 @ W2 + b2
a2 = sigmoid(z2)

loss = np.mean((a2 - y) ** 2)
print("loss:", loss)

dLoss_da2 = 2 * (a2 - y) / y.size

da2_dz2 = sigmoid_de(z2)
dz2_dW2 = a1

dLoss_dz2 = dLoss_da2 * da2_dz2

dW2 = a1.T @ dLoss_dz2
db2 = np.sum(dLoss_dz2, axis=0, keepdims=True)

dz2_da1 = W2.T
dLoss_da1 = dLoss_dz2 @ dz2_da1

da1_dz1 = sigmoid_de(z1)
dLoss_dz1 = dLoss_da1 * da1_dz1

dW1 = X.T @ dLoss_dz1
db1 = np.sum(dLoss_dz1, axis=0, keepdims=True)

W2 -= lr * dW2
b2 -= lr * db2
W1 -= lr * dW1
b1 -= lr * db1

print("Updated W1:\n", W1)
print("Updated W2:\n", W2)

#با کمی کمک از هوش مصنوعی
```
